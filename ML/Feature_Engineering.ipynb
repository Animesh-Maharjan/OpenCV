# Column Transformer

```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv("datasets/Exam_Score_Prediction.csv")
```

```python
df.head()
```

```python
df = df.drop("student_id", axis=1)
```

```python
df.isnull().sum()
```

```python
df["study_hours"] = df["study_hours"].fillna(df["study_hours"].median())
df["class_attendance"] = df["class_attendance"].fillna(df["class_attendance"].median())
df["sleep_hours"] = df["sleep_hours"].fillna(df["sleep_hours"].median())
df["exam_score"] = df["exam_score"].fillna(df["exam_score"].median())
```

```python
cols_to_clean = ["gender", "course", "study_method", "internet_access", "sleep_quality", "facility_rating","exam_difficulty"]

df.loc[:, cols_to_clean] = (
    df.loc[:, cols_to_clean]
    .apply(lambda x: x.str.lower().str.strip())
)
```

```python
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(
                                    df.drop("exam_score",axis=1),
                                    df["exam_score"],
                                    test_size=0.2,
                                    random_state=42
                                                 )
```

```python
oe = OrdinalEncoder(categories = [["poor", "average", "good"],
                                  ["low", "medium", "high"],
                                  ["easy", "moderate", "hard"],
                                  ["no","yes"]])
x_train_oe = oe.fit_transform(x_train[["sleep_quality","facility_rating","exam_difficulty","internet_access"]])

x_test_oe = oe.fit_transform(x_test[["sleep_quality","facility_rating","exam_difficulty","internet_access"]])

x_train_oe.shape
```

```python
# ohe = OneHotEncoder()
ohe = OneHotEncoder(drop = "first", sparse_output=False, dtype=int)

x_train_ohe = ohe.fit_transform(x_train[["gender", "course", "study_method"]])

x_test_ohe = ohe.fit_transform(x_test[["gender", "course", "study_method"]])

x_train_ohe.shape
```

```python
df.head()
```

```python
x_train_num = x_train.drop(columns = ["gender", "course", "study_method", "internet_access", "sleep_quality", "facility_rating","exam_difficulty"])
x_test_num = x_test.drop(columns = ["gender", "course", "study_method", "internet_access", "sleep_quality", "facility_rating","exam_difficulty"])
```

```python
x_train_transformed = np.concatenate((x_train_num, x_train_oe,x_train_ohe), axis = 1)
x_test_transformed = np.concatenate((x_test_num, x_test_oe,x_test_ohe), axis = 1)
```

```python
x_train_transformed.shape
```

Column transformer method

```python
from sklearn.compose import ColumnTransformer
```

```python
transformer = ColumnTransformer(transformers=[
    ("tnf1",OrdinalEncoder(categories = [["poor", "average", "good"],
                                        ["low", "medium", "high"],
                                        ["easy", "moderate", "hard"],
                                        ["no","yes"]]),
                                        ["sleep_quality","facility_rating","exam_difficulty","internet_access"]),
    ("tnf2", OneHotEncoder(drop = "first", sparse_output=False, dtype=int),["gender", "course", "study_method"]),
],remainder="passthrough")
```

```python
x_train_transformed = transformer.fit_transform(x_train)
```

```python
x_test_transformed = transformer.transform(x_test)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train_transformed)
x_test_scaled = scaler.transform(x_test_transformed)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor

# Linear Regression
lr = LinearRegression()
lr.fit(x_train_scaled, y_train)

# Or Random Forest
rf = RandomForestRegressor(random_state=42)
rf.fit(x_train_transformed, y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

y_pred = lr.predict(x_test_scaled)

print("MSE:", mean_squared_error(y_test, y_pred))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred)))
print("MAE:", mean_absolute_error(y_test, y_pred))
print("R² Score:", r2_score(y_test, y_pred))
```

```python
y_pred = rf.predict(x_test_transformed)

print("MSE:", mean_squared_error(y_test, y_pred))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred)))
print("MAE:", mean_absolute_error(y_test, y_pred))
print("R² Score:", r2_score(y_test, y_pred))
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(lr, x_train_transformed, y_train, cv=5, scoring='r2')
print(scores.mean())
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(rf, x_train_transformed, y_train, cv=5, scoring='r2')
print(scores.mean())
```
